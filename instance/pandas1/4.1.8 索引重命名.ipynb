# 4.1.8 索引重命名

```python
df = pd.read_excel('team.xlsx',index_col='name')
df.head()
```

```python
# 设置多层索引
df.set_index('Q4',append=True,inplace=True)
df.head()
```

```python
# 1，修改多层行索引名
df.rename_axis(['baidu','taobao'])
```

```python
# 2，修改多层行索引名
df.rename_axis(index={'name':'A','Q4':'B'})
```

```python
# 3，修改多层列索引名
df.rename(columns={'team':'kk','Q1':'BB'})
```

```python
# 4，将所有列名改为大写
df.rename(str.upper,axis='columns')
```
