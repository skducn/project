```python
import pandas as pd
```

# 4.1.2 建立索引

```python
df = pd.read_excel('team.xlsx')  
df.head()
```

```python
# 1，数据加载中指定索引，使用 index_col 参数
df = pd.read_excel('team.xlsx',index_col='name')  
df.head()
```

```python
# 2，数据加载中未指定索引，使用 df.set_index() 方法
# 单层索引 name
df2 = pd.read_excel('team.xlsx')
df2.set_index('name',inplace=True)
df2.head()
```

```python
# 3，两层索引 name、team
# 注意：索引过的列不能再索引，否则报错
df3 = pd.read_excel('team.xlsx')
df3.set_index(['name','team'],inplace=True)
df3.head()
```

```python
# 4，两层索引应用，姓名的第一个字母和姓名
df4 = pd.read_excel('team.xlsx')
df4.set_index([df4.name.str[0],'name'],inplace=True)
df4.head()
```

```python
# 5，将Series指定索引
df5 = pd.read_excel('team.xlsx')
s = pd.Series([i for i in range(100)])
df5.set_index(s,inplace=True)
df5.head()
```

```python
# 6，两层索引
df5.set_index([s,'name'],inplace=True)
df5.head()
```

```python
# 7，两层索引，计算索引s生成新的索引
df6 = pd.read_excel('team.xlsx')
df6.set_index([s, s**2],inplace=True)
df6.head()
```

```python
# 索引Q1
df7 = pd.read_excel('team.xlsx')
df7.set_index('Q1',inplace=True)
df7.head()
```

```python
# 8，保留原列Q2 ，即索引Q2的同时保留原列, 使用 drop=False 参数 
# 注意：此时上次Q1的索引没了，且原列Q1也没了。
df7.set_index('Q2',drop=False,inplace=True)
df7.head()
```

```python
# 9，保留原来的索引，即索引Q3的同时保留上次Q2索引，使用 append=True 参数
# 注意：上上次的Q1任然没
df7.set_index('Q3',append=True,inplace=True)
df7.head()
```

```python
df7.set_index('Q4',append=True, drop=False, inplace=True)
df7.head()
```
